```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.style as style 
```

```python
x = np.linspace(2003, 2020, 18)
label_color_list =  ["darkred"]+ ["darkgreen"]*15 + ["darkred", "darkgreen"] 
y = [0,0.44, 0.18, 0.43, 0.5, 0.12, 0.14, 0.27, 0.88, 0.38, 1.11, 0.56, 0.6, 1, 1, 0.86, 1.4, 0.67]
for i in x:
    i = int(i)
    i = str(i)
```

### Wykres poprawiony

```python

plt.style.use('_classic_test_patch')
fig = plt.figure(figsize=(9.75, 8))
ax = fig.add_axes([0,0,1,1])
ax.hlines(1, 2002.5, 2020.5, colors = "r", linestyles = "--", linewidth = 1)
ax.bar(x, height = y, width = 0.6, color = ['darkgreen']*16 + ["darkred"] + ["darkgreen"], alpha = 0.999)
ax.set_xticks(x)
ax.set_title("Liczba goli strzelonych przez Cristiano Ronaldo na mecz", fontsize=20)
ax.set_xlabel("Rok", fontsize=15)
ax.tick_params(axis='both', which='major', labelsize=13)
ax.set_ylabel("Liczba goli na mecz", fontsize=15)

rects = ax.patches
i = 0
for rect, label in zip(rects, y):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 0.002, label,
    ha='center', va='bottom', fontsize=14, color = label_color_list[i])
    i += 1;
plt.show()
```

Wykres poprawiony jest dużo bardziej czytelny i przyjaźniejszy dla oka. Zaznaczona w nim została wartość maksymalna oraz minimalna, wyraźnie zobrazowana sytuacja stabilna tzn. jeden gol na jeden mecz oraz poprawiona została ogólna szata graficzna wykresu, dzięki czemu odbiorca nie będzie znudzony jednolitością wykresu.
