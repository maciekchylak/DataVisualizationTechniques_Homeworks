```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.style as style 
```

```python
label = np.array(["Zdecydowanie dobrze", "Raczej dobrze", "Raczej źle", "Zdecydowanie źle", "Trudno powiedzieć"])
y = np.array([0, 0, 45, 52, 3])
```

### Wykres poprawiony

```python
plt.style.use('_classic_test_patch')
fig, ax = plt.subplots(figsize=(11, 9))
ax.bar(label, height = y, width = 0.6, color = "darkred", alpha = 0.999)
ax.set_title("Jak działania Kościoła w sprawie pedofilii oceniają ludzie młodzi (18-29 lat)?", fontsize=15)
ax.set_ylabel("Liczba opinii", fontsize=15)

rects = ax.patches
for rect, label in zip(rects, y):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 0.002, label,
    ha='center', va='bottom', fontsize=14)
plt.show()
```

Wykres "kwadratowy" nie jest najlepszym sposobem na przedstawianie wizualizacji danych, gdyż bardzo ciężko nam porównywać pola fragmentów kwadratów, a przypadek, w którym przerabiamy wykres kołowy na kwadratowy, nie zachowując proporcji pól jak w podanym przypadku, może być mylący dla części użytkowników. Dodatkowo ciężko dobrze odpowiedzieć na pytanie, co zrobić z argumentami (w przypadku wykresu kołowego/kwadratowego), dla których wartość jest zerowa, jednak są one integralną częścią ankiety. W podanym przykładzie autor luźno zamieścił te argumenty obok ankiety oraz zaznaczył, że wartości są równe zero. Warto zauważyć, że dużo lepszym pomysłem jest zastosowanie wykresu słupkowego, gdyż zdecydowanie się na niego automatycznie rozwiązuję oba wyżej wymienione problemy.
